# outliers

```python
for col in num_cols:
    fig = px.box(
        df,
        y=col,
        title=f"Boxplot of '{col}'",
    )
    fig.update_layout(yaxis_title=col)
    fig.show()
```

```python
numeric_cols = ['Traffic_Density', 'Speed_Limit', 'Number_of_Vehicles', 'Driver_Age', 'Driver_Experience']

for col in numeric_cols:
    fig = px.box(df, y=col, title=f'Outlier Detection in {col}')
    fig.show()
```

```python
outliers_report = {}

for col in numeric_cols:
    values = df[col].dropna()
    q1 = values.quantile(0.25)
    q3 = values.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = values[(values < lower_bound) | (values > upper_bound)]
    outliers_report[col] = {
        'Outlier Count': len(outliers),
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound
    }


import pandas as pd
pd.DataFrame(outliers_report).T
```

# DATA CLEANING

```python
object_cols = df.select_dtypes(include='object').columns
df[object_cols] = df[object_cols].fillna('unknown')
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(-1)
```

```python
df.isnull().sum()
```

```python
df_clean = df[
    (df['Time_of_Day'] != 'unknown') &
    (df['Weather'] != 'unknown') &
    (df['Road_Type'] != 'unknown') &
    (df['Road_Condition'] != 'unknown') &
    (df['Vehicle_Type'] != 'unknown') &
    (df['Road_Light_Condition'] != 'unknown') &
    (df['Accident_Severity'] != 'unknown') &
    (df['Driver_Age'] != -1) &
    (df['Driver_Experience'] != -1) &
    (df['Traffic_Density'] != -1) &
    (df['Speed_Limit'] != -1)
]
```

# EDA

```python
# Accidents by Road Type
fig = px.bar(df_clean, x='Road_Type', title='Accidents by Road Type')
fig.show()
```

```python
# Accident Severity by Time of Day
fig = px.histogram(df_clean, x='Time_of_Day', color='Accident_Severity',
                   barmode='group', title='Accident Severity by Time of Day')
fig.show()
```

```python
# Weather Conditions vs Accident Severity
fig = px.sunburst(df_clean, path=['Weather', 'Accident_Severity'],
                  title='Weather Conditions vs Accident Severity')
fig.show()
```

```python
# Driver Age vs Road Light Condition by Severity
fig = px.histogram(df_clean, x='Road_Light_Condition', y='Driver_Age',
             color='Accident_Severity',
             title='Driver Age vs Road Light Condition by Severity')
fig.show()
```

```python
# Accident Percentage by Time of Day
df['Accident_Flag'] = df['Accident_Severity'].apply(lambda x: 0 if pd.isna(x) or x == 'unknown' else 1)

accidents_by_time = df.groupby('Time_of_Day')['Accident_Flag'].sum().reset_index()
total_accidents = accidents_by_time['Accident_Flag'].sum()
accidents_by_time['Accident_Percentage'] = (accidents_by_time['Accident_Flag'] / total_accidents) * 100

order = ['Morning', 'Afternoon', 'Evening', 'Night']
accidents_by_time['Time_of_Day'] = pd.Categorical(accidents_by_time['Time_of_Day'], categories=order, ordered=True)
accidents_by_time = accidents_by_time.sort_values('Time_of_Day')

fig = px.line(accidents_by_time, x='Time_of_Day', y='Accident_Percentage', markers=True,
              title='Accident Percentage by Time of Day',
              labels={'Accident_Percentage': 'Accident Percentage (%)', 'Time_of_Day': 'Time of Day'})
fig.show()
```

```python
# Number of Accidents by Weather Condition
fig = px.bar(
    df_clean,
    x='Weather',
    title='Number of Accidents by Weather Condition',
    labels={'Weather': 'Weather Condition', 'count': 'Number of Accidents'},
)

fig.update_layout(
    xaxis_title='Weather Condition',
    yaxis_title='Number of Accidents',
    bargap=0.3
)

fig.show()
```

```python
# Number of Accidents by Road Condition
filtered_df = df_clean[
    (df_clean['Road_Condition'].str.lower() != 'unknown') &
    (df_clean['Accident_Severity'].str.lower() != 'unknown')
]

accidents_by_road_condition = filtered_df.groupby(
    ['Road_Condition', 'Accident_Severity']
).size().reset_index(name='Count')

fig = px.bar(
    accidents_by_road_condition,
    x='Road_Condition',
    y='Count',
    color='Accident_Severity',
    barmode='group',
    title='Number of Accidents by Road Condition ',
    labels={'Road_Condition': 'Road Condition', 'Count': 'Number of Accidents'}
)

fig.show()
```

```python
accidents_by_road_condition = df.groupby(['Road_Condition', 'Accident_Severity']).size().reset_index(name='Count')

fig = px.bar(accidents_by_road_condition, x='Road_Condition', y='Count', color='Accident_Severity',
             barmode='group', title='Number of Accidents by Road Condition',
             labels={'Road_Condition': 'Road Condition', 'Count': 'Number of Accidents'})
fig.show()
```

```python
#Average Speed Limit by Road Type
avg_speed_by_road = df.groupby('Road_Type')['Speed_Limit'].mean().reset_index()

fig = px.line(avg_speed_by_road, x='Road_Type', y='Speed_Limit', markers=True,
              title='Average Speed Limit by Road Type',
              labels={'Speed_Limit': 'Average Speed Limit', 'Road_Type': 'Road Type'})
fig.show()
```

```python
# Calculate the number of accidents for each vehicle type
accidents_by_vehicle = df_clean['Vehicle_Type'].value_counts().reset_index()
accidents_by_vehicle.columns = ['Vehicle_Type', 'Count']


fig = px.bar(
    accidents_by_vehicle,
    x='Vehicle_Type',
    y='Count',
    title='Number of Accidents by Vehicle Type',
    labels={'Vehicle_Type': 'Vehicle Type', 'Count': 'Number of Accidents'},
    color='Count',
    color_continuous_scale='Viridis'
)

fig.update_layout(
    xaxis_title='Vehicle Type',
    yaxis_title='Number of Accidents'
)

fig.show()
```
